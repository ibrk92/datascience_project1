### Splitting Dataset into Training and Test Sets

```python
from sklearn.model_selection import train_test_split
```

```python
features = df_v.drop(columns = ["price"])   ## independent variable
response = df_v["price"] ## dependent variable
features.head()
```

```python
## Importing 'train_test_split'
```

```python
## train_size = 0.90, test_size = 0.10
## random seed value = 1024
## splitting the data into training set and testing set
## features are my independent variables (this is without price)
## response are my dependent value (this is price)
X_train, X_test, Y_train, Y_test = train_test_split(features, response, test_size = 0.10, random_state = 1024)

print("Training and Testing are splitted successfully")
```

## Training and Testing

### A simple Linear Regression Model

```python
## try to understand that my dependent value(price-response-Y) has a relationship with independent variables (without price, features - X)
## importing library

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train,Y_train)
## model uses independent values (X_train) to predict dependent value (price)
## it generates parameters
```

```python
predictions = model.predict(X_test) ## we try to predict the dependent value using the independent values
```

```python
#predictions
```

### Calculate The Perfomance Metric

### R-squared

```python
## R-squared
r2_train = model.score(X_train, Y_train) ## it shows how our regression model predict the dependent variable 
print("R^2 in Training Set: ", r2_train)
r2_test = model.score(X_test, Y_test)  ## it shows how our regression model performs on the test data (in real data)
print("R^2 in Test Set: " , r2_test)
```

- R squared values is btw 0 and 1, if the value is close to 1, your model predict the data well
- The value of R^2 in Training set is 0.76519 so our model on training data predicts the dependent values(price) by 76.51%
- The value of R^2 in Test set is 0.78121 so our model on test data predicts the dependent values(price) by 78.12% (so the model generalizes well to real world data)


### RMSE

```python
from sklearn.metrics import mean_squared_error
```

```python
## calculation of mean squared error btw real-world-dependent value(Y_test) and predictions 
## The mean squared error calculate how far the predictions from the real data values
## So a smaller RMSE value indicates that the model makes better predictions.

mse = mean_squared_error(Y_test, predictions) 

print('RMSE is: ', mse**0.5) ##  we took the squares while calculating, we take the square root(**1/2) again so that we can see the real value.
```

### Linear Regression Model with Ridge

```python
#importing the libraries
from sklearn.linear_model import Ridge, Lasso 
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
alphas = 10**np.linspace(5,-2,10)
alphas
```

```python
## ridge = Ridge(normalize = True) normalize = True ies depreciated so i will do this manually with standartScaler
ridge = Ridge()
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
coefs = []

for a in alphas:
    ridge.set_params(alpha = a)
    ridge.fit(features,response)
    coefs.append(ridge.coef_)
    
np.shape(coefs)
```

```python
plt.plot(alphas, coefs)
ax = plt.gca()
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('weights')
```

```python
rmse_list = []
r2_list = []

for alpha in alphas:
    ridge = Ridge(alpha=alpha, max_iter=1000)
    ridge.fit(X_train_scaled, Y_train)
    pred = ridge.predict(X_test_scaled)
    
    r2_list.append(ridge.score(X_test_scaled, Y_test))
    rmse = mean_squared_error(Y_test, pred)**0.5
    rmse_list.append(rmse)
    print("Alpha:", alpha, "RMSE:", rmse)

ridge_result = np.vstack((alphas, rmse_list, r2_list)).T
ridge_df = pd.DataFrame(ridge_result, columns=['Alpha', 'RMSE', 'R2'])
print(ridge_df)
```

```python
#alpha = 12.915497 RMSE starts to increase and R2 starts to reduce. 
#It appears to provide a good balance between model fit and complexity.
ridge = Ridge(alpha = 12.915497)
ridge.fit(X_train_scaled, Y_train)             
pred = ridge.predict(X_test_scaled)           
print(pd.Series(ridge.coef_, index = features.columns))
```

### Polynomial Regression Model

```python
#importing the libraries
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Create polynomial features
poly_features = PolynomialFeatures(degree=2)  ## choosing the degree is 2, beacuse 
X_train_poly = poly_features.fit_transform(X_train)
```

```python
#Make predictions
X_test_poly = poly_features.transform(X_test)
Y_pred = poly_reg.predict(X_test_poly)
```

```python
# Calculate R-squared (R2)
poly_r2_train = poly_reg.score(X_train_poly, Y_train) 
print("R^2 in Training Set: ", poly_r2_train)
poly_r2_test = poly_reg.score(X_test_poly, Y_test)
print("R^2 in Test Set: " , poly_r2_test)

# Calculate Root Mean Squared Error (RMSE)
poly_mse = mean_squared_error(Y_test, Y_pred) 
print('RMSE is: ', poly_mse**0.5)
```

```python
#Plot residual values
residuals = Y_test - Y_pred
plt.scatter(Y_pred, residuals)
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='-')
plt.title("Residual Plot")
plt.show()
```

```python
#Visualize coefficient
coef = poly_reg.coef_
plt.plot(coef)
plt.xlabel("Feature Index")
plt.ylabel("Coefficient Value")
plt.title("Polynomial Regression Coefficients")
plt.show()
```
